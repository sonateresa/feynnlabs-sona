```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
ev_data_2_cleaned = ev_data_2[['latitude', 'longitude']].dropna()

coordinates = ev_data_2_cleaned[['latitude', 'longitude']]
```

```python
print(ev_data_2['latitude'].unique())
print(ev_data_2['longitude'].unique())
print(ev_data_2['longitude'].dtype)
```

```python
scaler = StandardScaler()
coordinates_scaled = scaler.fit_transform(coordinates)
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(coordinates_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
optimal_k = 3
```

```python
# Perform KMeans clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
ev_data_2_cleaned['Cluster'] = kmeans.fit_predict(coordinates_scaled)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', data=ev_data_2_cleaned, hue='Cluster', palette='viridis', s=80)
plt.title('K-means Clustering of Charging Stations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Cluster')
plt.show()
```
